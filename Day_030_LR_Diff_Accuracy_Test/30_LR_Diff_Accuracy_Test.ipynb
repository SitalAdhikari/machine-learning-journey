```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df = pd.read_csv('dataset/placement_cgpa_package.csv')
df.sample()
```

```python
X= df.iloc[:,0].values
y = df.iloc[:,-1].values
X_train,X_test,y_train,y_test = train_test_split(X.reshape(-1,1),y.reshape(-1,1),random_state=42, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
print('MAE', mean_absolute_error(y_test,y_pred))
#MAE is doing error of 0.23 lpa
```

```python
print('MSE', mean_squared_error(y_test,y_pred))
# MSE is doing error of 0.084 square lpa
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
##RMSE is doing error of 0.29 lpa
```

```python
r2 = r2_score(y_test,y_pred)
print('R2:',r2)
```
