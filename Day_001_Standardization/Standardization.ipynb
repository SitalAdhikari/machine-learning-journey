```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
df.sample(5)
```

```python
df.sample(5)
```

##### Train test split

```python
df=df.iloc[:,1:]
```

```python
from sklearn.model_selection import train_test_split
#X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:,1:3], df.iloc[:,-1], test_size=0.3, random_state=0)
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('placement', axis=1), df['placement'], test_size=0.3, random_state=42)
```

#### Standard Scaler  - standardization

```python
X_train
```

```python
Y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
# after stardardscaling we will get numpy array
# hence converting numpy arrays into pandas dataframe

X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
#without columns we will get 0 and 1 as columns , columns just give name to the column
```

```python
# X_train.describe() -> gives around 7 8 digits after decimal
# hence we use the round 
np.round(X_train_scaled.describe(),1)
```
