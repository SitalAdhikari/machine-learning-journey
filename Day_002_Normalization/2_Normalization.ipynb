# <center> Normalization </center>

##### Normalization is a feature scaling technique that transforms data so that all the feature values lie in the specific range.

#### <u>MinMax Scaling </u>

##### - Range -> [0,1]


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('placement.csv')
data.sample(3)
```

```python
data = data.iloc[:,1:]
data.sample(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, Y_test = train_test_split(data.drop('placement',axis=1),data.placement, test_size=0.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled,  columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns= X_test.columns)

np.round(X_train_scaled.sample(3),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

##### <br> <u><h4> Mean Normalization </h4></u>
Mean Centering - 0     <br>
Range-> [-1,1] <br>
We have to use this if we need centered data. Alternatively, for center data, we can use standardization also.
Hence, there is minimal use of mean normalization.

##### <br> <u><h4> MaxAbs Normalization </h4></u>
If there is high number of zeros in the input, then it is better to use MaxAbs Normalization.

##### <br> <u><h4> Robust Normalization </h4></u>
If your data has a lot of outliers, then it is said that using robust normalization will give high accuracy result.

### Normalization vs Standardization
<br>
1) Is feature scaling required?-> If you are working with decision tree, feature scaling won't be required.<br>
2) Mostly used Standardization. <br>
3) MinMax Scaling: if you know min and max value then use it like image processing, CNN [0-255]
