### Feature Construction

```python
X.sample()
```

```python
X['Family_size']= X['sibsp'] +X['parch']+1
```

```python
X.sample()
```

```python
def famfunc(num):
    if (num == 1):
        return 0 #alone
    elif(num >1 and num<=4):
        return 1 #small family
    else:
        return 2 #large family
```

```python
X['Family_type']=X['Family_size'].apply(famfunc)
```

```python
X.head()
```

```python
X = X.drop(columns=['sibsp','parch','Family_size'])
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y.values.ravel(),scoring='accuracy',cv=20))
```

### Feature Splitting

```python
df = pd.read_csv('dataset/titanic.csv', usecols=['name'])
df.sample()
```

```python
#let's split the name
# we use .split() fn
# .split() returns splitted values in list in same column
# to make different columns .split( ,expand =True)
```

```python
df[['Surname','Remaining_name']] = df['name'].str.split(',', expand=True)
df.sample()
```

```python
df[['Title','First_name']] = df['Remaining_name'].str.split('.',n=1, expand=True)
df.sample(5)
```
