```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
df = pd.read_csv('dataset/titanic_dataset.csv', usecols=['Age','Pclass','Fare','Survived'])
```

```python
df.sample(5)
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop('Survived',axis=1)
y=df.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size=0.2)
```

```python
X_train.head()
```

```python
knn= KNNImputer()
X_train_imputed = knn.fit_transform(X_train)
X_test_imputed = knn.transform(X_test)
```

```python
pd.DataFrame(X_train_imputed, columns=X_train.columns).sample(5)
```

```python
X_train_imputed
```

```python
lr = LogisticRegression()
lr.fit(X_train_imputed,y_train)
y_pred = lr.predict(X_test_imputed)
accuracy_score(y_pred,y_test)
```

```python
#comparing with mean imputation using simple imputation
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)

lr= LogisticRegression()
lr.fit(X_train_trf,y_train)
y_pred=lr.predict(X_test_trf)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer
```

```python
data = {
    'age':[25,30,np.nan , 40,35],
    'salary':[50000,np.nan,60000,80000,np.nan],
    'experience':[2,5,7,np.nan,10]
}
df = pd.DataFrame(data)
df
```

```python
impute = IterativeImputer(max_iter=10,random_state=42)
df_imputed = impute.fit_transform(df)
df_imputed = pd.DataFrame(df_imputed,columns=df.columns)
np.round(df_imputed,2)
```
