#### Bar Graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('penguins')
```

```python
sns.barplot(x=df.species ,y= df.body_mass_g)
# sns.barplot(x='species', y='body_mass_g',data=df)

plt.show()
```

```python
#hue - do differentiate the male and female
#order - to change the position of the portion of the species
# hue_order=['Female', 'Male'] - to put the female portion before the male portion
#errorbar=('ci',95) (0-100) ---- error bar length
#n_boot = 3   ---> research yourself (hw)

#horizontal barplot - 2 axes should be in numeric (single one is not in the alphabets)
#sns.barplot(x='hello',y='billo',data=df,orient='h')
#vorizontal barplot (below fig - default- orient='v')


import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('penguins')
order1=['Chinstrap','Gentoo','Adelie']
sns.barplot(x='species', y='body_mass_g',data=df , hue='sex', order = order1 , hue_order=['Female', 'Male'],errorbar=('ci',100), n_boot=100)
plt.show()
```

```python
#saturation - to change the color's contrast
#err_kws = {'color':'red','linewidth':2} -> color and errorbar width change for the error bar 
#capsize -> horizontal size of the two errorbar bounded lines at the top

import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("penguins")
sns.barplot(x='species', y='body_mass_g',data=df, saturation=0, capsize=0.2 ,  err_kws = {'color':'red','linewidth':1 })
plt.show()
```

```python
#dodge -> overlap two portions
#alpha -> to make the graph more light

#sns.set(style='darkgrid')   ->  {... = darkgrid(for dark grid) , dark(dark background),...}
sns.barplot(x='island',y='bill_length_mm',data = df, hue='sex',alpha=0.5,dodge= False)
plt.show()
```

#### Heat Map

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
var = np.linspace(2,20,10).reshape(2,5)
sns.heatmap(var)
plt.show()
```

```python
#  vmax  , vmin  ->  we can change the maximum and minimum value of the index of the heatmap
# cmap  -> change the color of the index of the heatmap
# annot -> show the value of the portion in the color box


sns.heatmap(var, vmax=25, vmin=6, cmap="PuOr" , annot= True)
plt.show()
```

```python
# annot is used and fmt should be specified  s for string  ->giving our own new data in the heatmap portion
# change the fontsize and the color of the font:
# linewidth -> to get the gap between two  portion
#linecolor -> to change the color of the linespace
#cbar -> the index bar of the right side of the heatmap
#xticklabels= False -> to remove the label of the x-axis
#yticklabels=False -> to remove the label of the y-axis

var = np.linspace(1,20,4).reshape(2,2)
print(var)

font={"fontsize": 20 , "color": "r"}
array1 = np.array([["a1", "a2"],
                 ["a3","a4"]])

sns.heatmap(var, annot = array1 , fmt="s" , annot_kws = font, linewidth=4, linecolor="y", cbar=False , xticklabels=False)
plt.show()
```

#### Pair Plot

```python
df = sns.load_dataset("tips")
df
```

```python
#sns.pairplot(df)
sns.pairplot(df, hue="sex")
plt.show()
```

```python
sns.pairplot(df, hue="sex", vars=["total_bill","tip"] , hue_order=['Female','Male'] , palette='plasma_r')
plt.show()
```

```python
# x_vars -> to change the position of the total _bill and tip in the x-axis we can use x_vars ,, for y-axis we can use the y-axis
# kind="reg"  -> for linear line in the points of the graphs. [reg,kde,hist,scatter]
#diag_kind="hist" -> give the hist only in the diagonal parts
sns.pairplot(df, hue="sex",x_vars=["tip","total_bill"] , hue_order=['Female','Male'] , palette='plasma_r', kind='reg')
plt.show()
```

```python
sns.pairplot(df, hue="sex", markers=['^','*'])
plt.show()
```
