```python
#seaborn needs dataframe to built the graphs (unlike Matplotlib)

df=pd.DataFrame({"data1":[1,2,3,4,5] , "data2":[1,2,3,4,5]})

sns.lineplot(x='data1',y='data2' ,data= df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset = sns.load_dataset('planets')
#legend is given, if we want to remove the legend then we can do legend=False
sns.lineplot(x='mass',y='orbital_period',data=dataset, hue="year", legend=False)
plt.grid()
plt.title("Planet's Rotations")


plt.show()
```

#### Bar Graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('penguins')
```

```python
sns.barplot(x=df.species ,y= df.body_mass_g)
# sns.barplot(x='species', y='body_mass_g',data=df)

plt.show()
```

```python
#hue - do differentiate the male and female
#order - to change the position of the portion of the species
# hue_order=['Female', 'Male'] - to put the female portion before the male portion
#errorbar=('ci',95) (0-100) ---- error bar length
#n_boot = 3   ---> research yourself (hw)

#horizontal barplot - 2 axes should be in numeric (single one is not in the alphabets)
#sns.barplot(x='hello',y='billo',data=df,orient='h')
#vorizontal barplot (below fig - default- orient='v')


import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('penguins')
order1=['Chinstrap','Gentoo','Adelie']
sns.barplot(x='species', y='body_mass_g',data=df , hue='sex', order = order1 , hue_order=['Female', 'Male'],errorbar=('ci',100), n_boot=100)
plt.show()
```

```python
#saturation - to change the color's contrast
#err_kws = {'color':'red','linewidth':2} -> color and errorbar width change for the error bar 
#capsize -> horizontal size of the two errorbar bounded lines at the top

import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("penguins")
sns.barplot(x='species', y='body_mass_g',data=df, saturation=0, capsize=0.2 ,  err_kws = {'color':'red','linewidth':1 })
plt.show()
```

```python
#dodge -> overlap two portions
#alpha -> to make the graph more light

#sns.set(style='darkgrid')   ->  {... = darkgrid(for dark grid) , dark(dark background),...}
sns.barplot(x='island',y='bill_length_mm',data = df, hue='sex',alpha=0.5,dodge= False)
plt.show()
```

#### Histogram plot

```python
import pandas as pd
import matplotlib.pyplot as pd
import seaborn as sns

df=sns.load_dataset('penguins')
```

```python
sns.displot(df['flipper_length_mm'])
plt.show()
```

```python
#bins for the proper labeling of the x-axis numbers
#kde = True  -> kernel density estimator -> it estimate the point and join the points and make the curve
#rug= True -> small line labelling in the x-axis 

sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230],kde=True, rug=True ,color='red')
plt.show()
```

```python
#log_scale = True -> convert the data of x-axis in the log value and show the graph

sns.displot(df['flipper_length_mm'], log_scale=True)
plt.show()
```
