## Example of matplotlib

```python
x=[10,20,30,40]
y=[20,15,25,20]

plt.plot(x,y) #we can use plot to make the line graph (it will make the graph)
plt.show() # it will show the graph made by plt.plot()
```

### Some Key terms:

#####  
1.Plot = Final name of the diagram (curve, line) 
2.Line style = .... , ------ , -.-.
3.Legend =  (eg: red color represent the Number of Sales )
4. Label = name of the ghes
5. Grid = small boxes 
6.Funtion = (eg: plot(), show())
7. Parameters = Variables
8. DPI = dot per inch (quality of the picture)

```python
import matplotlib.pyplot as plt

x = ['Sun','Mon','Tues','Wednes',"Thrus",'Fri']
y= [5000,7000,1000,2000,3000 ,2000]

plt.plot(x,y, color = 'red', linestyle='--', linewidth='1.5', marker='^', label='trend') # after x,y -- others are optionals
#marker: o= circle , s =square , d = diamond, x = cross, + = + , ^ = triangle

#plt.legend() --- to show the label which symbol represents what in the top right corner
plt.legend(loc='upper left')  # plt.legend(loc='location') 
#location is optional -  upper left / lower right / upper right / lower left default(upper right)

plt.grid() # plt.grid( color='gray' , linestyle=':', linewidth='2')

plt.title('Sales per day')
plt.xlabel('Days')
plt.ylabel('Total Sell (USD)')

plt.show()  # to display the graph
```

#### Limit in the ghes

```python
import matplotlib.pyplot as plt
x = ['Sun','Mon','Tues','Wednes',"Thrus",'Fri']
y= [5000,7000,1000,2000,3000 ,2000]
plt.plot(x,y, color = 'red', linestyle='--', linewidth='1.5', marker='o', label='trend')
plt.legend(loc='upper left')
plt.grid() 

#ghis limit
plt.xlim('Sun', 'Thrus') #to limit in the x-ghis
plt.ylim(0,5000) # to limit in the y-ghis


plt.show() 
```

### 2 lines in same graphs

```python
import matplotlib.pyplot as plt
x = ['Sun','Mon','Tues','Wednes',"Thrus",'Fri']
y= [5000,7000,1000,2000,3000 ,2000]
plt.plot(x,y, color = 'red', linestyle='--', linewidth='1.5', marker='o', label='trend1')
plt.plot(x,[2000,3000,4000,5000,3000,4500],color='green' , label='trend2')
plt.legend(loc='upper right')
plt.grid() 
plt.show() 
```

## Bar Graph - plt.bar()

```python
import matplotlib.pyplot as plt
x = ['Sun','Mon','Tues','Wednes',"Thrus",'Fri']
y= [5000,1200,1000,2000,3000 ,2000]

plt.bar(x,y)  #plt.barh() for the horizontal lines in the bargraph
#plt.bar(x,y,color='red', label='week1')

plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales Per Day')
plt.show()
```

## Pie Chart - plt.pie()

```python
import matplotlib.pyplot as plt
mobile=['Iphone','Samsung','Nothing']
sale=[3000,7000,1000]

#plt.pie(sale,labels=mobile)
plt.pie(sale,labels=mobile, colors=['red','skyblue','lightgreen'], autopct='%1.1f%%')
plt.show()

```

## Histogram (joined bargraph under certain condition(x,f)) -plt.hist()

```python
import matplotlib.pyplot as plt
weight=[67,87,56,55,54,76,45,35,76,56,76,87,55,44,77,65,45,66,68,76,77]

#plt.hist(weight, bins=5) bins means interval 
plt.hist(weight,bins=5,color='black',edgecolor='white')
plt.title('Weight of the Students of Undergraduate')
plt.xlabel('Weight of the students')
plt.ylabel('Frequency')
plt.show()
```

## Scatter Points

```python
import matplotlib.pyplot as plt
stall=[3,4,5,6,7]
sales=[5000,5500,5800,6000,6100]

#plt.scatter(stall,sales)
plt.scatter(stall,sales, color='green',marker='o',label="sales")
plt.legend()
plt.grid()
plt.xlabel('Number of shops')
plt.ylabel('Total Sales')
plt.title('Number of shops vs Total sales')
plt.show()
```

## Combining the different charts and graphs in the same page, so that we can analyze and understand the data properly 


##### Example: taking 3 graphs of the saving, income, expenditure together can help us to predict the future family fund management

#### 1.Subplot 

```python
import matplotlib.pyplot as plt

#plt.subplot(rows,colums,index)   
plt.subplot(1,2,1) #1 row, 2 column, 1 subplot
x = ['Sun','Mon','Tues','Wednes',"Thrus",'Fri']
y= [5000,7000,1000,2000,3000 ,2000]
plt.plot(x,y, color = 'red', linestyle='--', linewidth='1.5', marker='o', label='trend1')
plt.plot(x,[2000,3000,4000,5000,3000,4500],color='green' , label='trend2')
plt.legend(loc='upper right')
plt.grid()

plt.subplot(1,2,2) #1 row, 2 column, 2 subplot
stall=[3,4,5,6,7]
sales=[5000,5500,5800,6000,6100]
plt.scatter(stall,sales, color='green',marker='o',label="sales")
plt.legend()
plt.grid()
plt.xlabel('Number of shops')
plt.ylabel('Total Sales')
plt.title('Number of shops vs Total sales')

plt.show()
```

#### 2. Object Oriented (recommended than subplot) - we can manipulate its size

```python
import matplotlib.pyplot as plt

x = [5,4,3,2,1]
y = [5000,4000,3000,2000,1000]

fig, ax= plt.subplots(1,2,figsize=(10,5))

#ax[0].plot(x,y)
ax[0].plot(x,y, color='red')
ax[0].set_title('linegraph')

ax[1].scatter(x,y)
ax[1].set_title('scatter points')

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [5,4,3,2,1]
y = [5000,4000,3000,2000,1000]
fig, ax= plt.subplots(1,2,figsize=(10,5))
 
fig.suptitle('Comparision') # to give the title of the all the graphs at the top

ax[0].plot(x,y)
ax[0].set_title('linegraph')
ax[1].scatter(x,y)
ax[1].set_title('scatter points')

plt.tight_layout() # to cover the whole dashboard

plt.show()
```
