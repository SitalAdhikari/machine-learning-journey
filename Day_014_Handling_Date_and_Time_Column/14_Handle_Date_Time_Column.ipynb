#### Handling dates

```python
# 1. converting info datetime type
date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
# 2. Extract Year
date['date_year']=date['date'].dt.year
date.sample(5)
```

```python
#extract months
date['date_month_no'] = date['date'].dt.month
date.sample(5)
```

```python
# month name
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python
#extract days
date['date_day']= date['date'].dt.day
date.sample(5)
```

```python
#week (0,1,2,3,4,5,6)
date['date_week'] = date['date'].dt.dayofweek
date.sample(5)
```

```python
date.drop(columns=['product_id','city_id'], inplace= True)
# day name - sunday, monday
date['date_day_name'] = date['date'].dt.day_name()
date.sample(5)

# starts from monday  0- Monday , 1- Tuesday , ....,6- Sunday
```

```python
# is weekend or not??
date['weekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
date.sample(5)
```

```python
# week of the year
date['date_year_week']= date['date'].dt.isocalendar().week
date.sample(5)
```

```python
date.drop(columns=['date_month_no','date_month_name','date_day','date_week'], inplace= True)
#quater of the year
date['quarter']=date['date'].dt.quarter
#semester of the year
date['semester']= np.where(date['quarter'].isin([1,2]),1,2)
date.sample(5)
```
